Se distribuye las funciones en base al orden de lista, con 100 números de ejemplo. 
En un local de comida rapida se requiere saber el total de ventas en cada semana, tomando en cuenta que segun sus datos registrado en su registro de ventas, nos dice que: en cada semana se vende un valor minimo de 400 dolares y un maximo de 650 dolares pero por lo general, en el registro consta que en la mayoria de las semanas se venden unos 530 dolares. Mediante una distribución triangular simular el comportamiento del mismo.

```python
#importamos las libreias necesarias.
import numpy as np
import random
import matplotlib.pyplot as plt
import math
```

```python
#Creamos un metodo para generar numero pseudoaleatorios
#Metodo generar numero con random de python
import random
def random_python(N):
    numeros = []
    random.seed(1032022)
    [numeros.append(random.random())for i in range(N)]
    return numeros
```

```python
#Llamo al metodo generar numeros
print("NUMEROS RANDOM")
numeros= random_python(100)
print(numeros)
```

```python
# Aplicacion de la Funcion Triangular
a = 530
contador = 0
valor=0
ventas = []
ventassemana= []
print("Semanas  \t ri \t probabilidad ")
for i in numeros:
    valor = a*(1- math.sqrt(1 - i))
    ventas.append(valor)
    contador = contador + 1
    ventassemana.append(contador)
    print(str(contador) + "\t" + str(i) + "\t"+ str(valor))
```

```python
# Funcion de Densidad
valorDensidad = []
contador2 = 0
print("Semanas  \t ri \t Valor de Densidad")
for i in numeros:
    valorD = 2/a * (1 - i/a)
    valorDensidad.append(valorD)
    contador2 = contador2 + 1
    print(str(contador2) + "\t" + str(i) + "\t"+ str(valorD))
```

```python
#graficamos
x = ventassemana
y = ventas
plt.title('Probabilidad Acumulada')
## Legenda en el eje y
plt.xlabel('semanas')
## Legenda en el eje x
plt.ylabel('ventas')
plt.plot(x,y)
## Mostramos Gráfica
plt.show()
```

```python
#graficamos
x = ventassemana
y = valorDensidad
plt.title('Densidad')
## Legenda en el eje y
plt.xlabel('semanas')
## Legenda en el eje x
plt.ylabel('densidad')
plt.plot(x,y)
## Mostramos Gráfica
plt.show()
```
